```python
extension = ''
for n in range(100):
    cow = n % 10
    dec = n // 10
    if dec != 1:
        if cow == 1:
            extension = 'a'
        elif 2 <= cow <= 4:
            extension = "y"
    else:
        extension = ""
    # print(n, "korov" + extension)
    print(n, extension, sep="    ")
```

```python
extension = ''
n = int(input())
cow = n % 10
if 5 <= n <= 20:
    extension = ""
else:
    if cow == 1:
        extension = 'a'
    elif 2 <= cow <= 4:
        extension = "y"
    elif cow == 0:
        extension = ""

print(n, "korov" + extension)
```
